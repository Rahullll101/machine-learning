```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

what is matrices?


it is model evaluation technique present in sckit-learn 

```python
df = pd.DataFrame({
    'study_hours': list(range(1, 11, 1)),
    'exam_score': list(range(10, 101, 10))
})
df.head()

# CONCLUSION: Created a perfect linear dataset showing the relationship between study hours (1-10) and exam scores (10-100), displaying first 5 rows for verification
```

```python
x = df[['study_hours',]]
y =  df['exam_score']

# CONCLUSION: Separated independent variable (study_hours) as 2D array and dependent variable (exam_score) as 1D array for supervised learning
```

since x is independent variable it should be 2d array, 
since y is independent variable it should be 1d array


```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=42)
x_train,x_test,y_train,y_test

# CONCLUSION: Split data into 80% training and 20% testing sets with fixed random state for reproducible model evaluation
```

    create model

```python
model = LinearRegression()

# CONCLUSION: Created a Linear Regression model instance ready for training on study hours vs exam score relationship
```

```python
# fit or train the model

model.fit(x_train,y_train)

# CONCLUSION: Successfully trained the model to learn how study hours affect exam scores using the training data
```

```python
model.intercept_

# CONCLUSION: Retrieved the y-intercept representing the expected exam score when study hours is zero
```

```python
# Returns the coefficient(s) of the trained linear regression model, representing the relationship between study_hours and exam_score
model.coef_

# CONCLUSION: Retrieved the slope coefficient showing how much exam score increases per additional study hour
```

```python
model.predict(x_test)

# CONCLUSION: Generated exam score predictions for test study hours to evaluate model performance
```

```python
y_pred = model.predict(x_test)

# CONCLUSION: Stored predictions in variable for further analysis and comparison with actual exam scores
```

```python
y_test - y_pred

# CONCLUSION: Calculated residuals (prediction errors) to understand how accurately the model predicts exam scores
```

```python
plt.scatter(x_test,y_test)

# CONCLUSION: Visualized actual exam score data points to see the true relationship between study hours and scores
```

```python
plt.scatter(x_test,y_pred)

# CONCLUSION: Visualized predicted exam scores to compare with actual data and assess model fit quality
```

```python
r2_score(y_test,y_pred)  # checks r2_score between actual outcome valueand predicted outcome value

# CONCLUSION: Calculated R² score to measure how well study hours explain the variance in exam scores (closer to 1 means perfect fit)
```

```python
model.score(x_test,y_test)
'''    The code `model.score(x_test, y_test)` calculates the R² score (coefficient of determination) of the linear regression model on the test data. It measures how well the model's predictions match the actual values; a score of 1.0 means perfect prediction.'''

# CONCLUSION: Alternative method to get R² score directly from the model - confirms model performance assessment
```

```python
# model.score vs r2_score:
# Both model.score and r2_score return the coefficient of determination (R^2 score) for regression models.
# model.score(X, y) is a built-in method of scikit-learn estimators and computes the R^2 score using the model's predict method.
# r2_score(y_true, y_pred) is a standalone function that requires you to provide the true and predicted values.
# In practice, model.score(X_test, y_test) is equivalent to r2_score(y_test, model.predict(X_test)).
```

```python
mean_squared_error(y_test,y_pred)

# CONCLUSION: Calculated Mean Squared Error to measure average squared difference between actual and predicted exam scores
```

```python
mean_absolute_error(y_test,y_pred)

# CONCLUSION: Calculated Mean Absolute Error to measure average absolute difference between actual and predicted scores (easier to interpret than MSE)
```

```python
# The difference between mean_absolute_error and mean_squared_error:
# - mean_absolute_error (MAE) calculates the average of the absolute differences between actual and predicted values. It gives equal weight to all errors.
# - mean_squared_error (MSE) calculates the average of the squared differences between actual and predicted values. It penalizes larger errors more than smaller ones due to squaring.
# In summary, MSE is more sensitive to outliers than MAE.
```
