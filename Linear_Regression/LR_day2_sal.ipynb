```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

```

```python
df = pd.read_csv('Salary_dataset.csv')

# CONCLUSION: Successfully loaded the salary dataset into a DataFrame for analysis
```

```python
df.head()

# CONCLUSION: Displayed first 5 rows to understand the dataset structure and verify successful data loading
```

```python
df

# CONCLUSION: Displayed the complete dataset to examine all records and identify any data quality issues
```

```python
df.duplicated().any()

# CONCLUSION: Checked for duplicate rows to ensure data quality and avoid bias in model training
```

```python
df=df.drop(columns='Unnamed: 0')

# CONCLUSION: Removed unnecessary index column to clean the dataset for analysis
```

```python
df.info()

# CONCLUSION: Examined dataset structure, data types, and memory usage to understand the data characteristics
```

```python
df.describe()

# CONCLUSION: Generated statistical summary (mean, std, min, max, quartiles) to understand data distribution and identify outliers
```

```python
x = df[["YearsExperience"]]
y = df['Salary']

# CONCLUSION: Separated independent variable (Years Experience) and dependent variable (Salary) for supervised learning
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=42)
x_train,x_test,y_train,y_test

# CONCLUSION: Split data into 80% training and 20% testing sets with fixed random state for reproducible results
```

## Create Model

```python
model = LinearRegression()

# CONCLUSION: Created a Linear Regression model instance ready for training on salary prediction task
```

```python
# fit or train the model

model.fit(x_train,y_train)

# CONCLUSION: Successfully trained the model to learn the relationship between years of experience and salary
```

```python
model.intercept_

# CONCLUSION: Retrieved the y-intercept representing the base salary when years of experience is zero
```

```python
model.coef_

# CONCLUSION: Retrieved the slope coefficient showing how much salary increases per year of experience
```

```python
model.predict(x_test)

# CONCLUSION: Generated salary predictions for test data to evaluate model performance
```

```python
y_pred = model.predict(x_test)
y_pred

# CONCLUSION: Stored predictions in variable for further analysis and comparison with actual salaries
```

```python
y_test - y_pred

# CONCLUSION: Calculated residuals (prediction errors) to understand how far off our predictions are from actual salaries
```

```python
plt.scatter(x_test,y_test)

# CONCLUSION: Visualized actual salary data points to see the true relationship between experience and salary
```

```python
plt.scatter(x_test,y_pred)

# CONCLUSION: Visualized predicted salary values to compare with actual data and assess model fit quality
```

```python
r2_score(y_test,y_pred)  # checks r2_score between actual outcome valueand predicted outcome value

# CONCLUSION: Calculated R² score to measure how well the model explains the variance in salary data (closer to 1 is better)
```

```python
model.score(x_test,y_test)

# CONCLUSION: Alternative way to get R² score directly from the model object for quick performance assessment
```

```python
mean_squared_error(y_test,y_pred)

# CONCLUSION: Calculated Mean Squared Error to measure average squared difference between actual and predicted salaries
```

```python
mean_absolute_error(y_test,y_pred)

# CONCLUSION: Calculated Mean Absolute Error to measure average absolute difference between actual and predicted salaries (easier to interpret than MSE)
```
